```python
from bs4 import BeautifulSoup
import requests
```

```python
#Task 1
url = 'https://www.theguardian.com/books/2022/jan/24/the-big-idea-should-animals-have-the-same-rights-as-humans'
req = requests.get(url)
source = req.text
soup = BeautifulSoup(source, 'html.parser')
links = soup.find_all('a',attrs={'data-link-name':'in body link'})
for link in links:
    print(link['href'], link.text)
```

```python
url = 'https://www.theguardian.com/uk/technology'
req = requests.get(url)
source = req.text
soup = BeautifulSoup(source, 'html.parser')
tech_div = soup.find('div',attrs={'data-title':'Technology'}).find_all('a', attrs={'class': 'js-headline-text'})
for article in tech_div:
    print(article['href'],article.text)
# tech_soup = BeautifulSoup(tech_div)
# articles = tech_soup.find_all('a', attrs={'class': 'js-headline-text'})
# print(articles)
```

```python
#Task 4
url = 'https://www.theguardian.com/technology?page='
articles = []
page = 1
while len(articles) < 50:
    req = requests.get(url+str(page))
    source = req.text
    soup = BeautifulSoup(source, 'html.parser')
    links = soup.find_all('a', attrs={'data-link-name':"article",'class':"fc-item__link"})
    for link in links:
        articles.append(link)
    page += 1
    
articles = articles[:50]

for article in articles:
    print(article['href'],article.text)
```

```python
#Task 5
import json
url = 'https://content.guardianapis.com/technology?api-key=4f8d263b-4da6-4d98-9734-bc38cc74a556'
req = requests.get(url)
src = req.text 
technology = json.loads(src)['response']
print(technology.keys())
print(json.dumps(technology['results'], indent=2, sort_keys=True))
```

```python
# Specify the arguments
args = {
    'section': 'technology', 
    'order-by': 'newest', 
    'api-key': '4f8d263b-4da6-4d98-9734-bc38cc74a556', 
    'page-size': '100',
    'q' : 'privacy%20AND%20Whatsapp%20OR%20privacy%20AND%20Signal'
}

# Construct the URL
base_url = 'http://content.guardianapis.com/search'
url = '{}?{}'.format(
    base_url, 
    '&'.join(["{}={}".format(kk, vv) for kk, vv in args.items()])
)

# Make the request and extract the source
req = requests.get(url) 
src = req.text
response  = json.loads(src)['response']
print(json.dumps(response['results'], indent=2, sort_keys=True))
```
